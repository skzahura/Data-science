```python
# which image loading function to use?
import time
def test_read_image(imgfile, func): #read the image with the function, image from the path given
    t0 = time.time() #time library to measure time
    img = func(imgfile)
    t = time.time() - t0
    return t

functions = [io.imread, Image.open, skimage.io.imread, cv2.imread]

times = []
for func in functions:
    ts = [] #create list of our times
    
    # load each image
    for f in filenames:
        name = f.split(".")[0]
        t = test_read_image(os.path.join(path, name + ".png"), func)
        ts.append(t)
    times.append(ts)
```

```python
# Note: we are calculating the mean time for each image loading method here
for t, func in zip(times, functions):
    print(str(func))
    print("{:.6f} seconds".format(np.mean(t)))
    print("_____")
```

### Calculate [PSNR](https://www.ni.com/de-de/innovations/white-papers/11/peak-signal-to-noise-ratio-as-an-image-quality-metric.html) between images





```python
# psnr means peak segment to noise ratio, how much noise between images if we compare
```

```python
def psnr(img1, img2):
    #mean squared error:
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return "MSE is zero" #compare m with n matrix each other and calculate distance between them, we dont need for loop
    PIXEL_MAX = 255.0 #MAX from the function/formula
    return 20 * np.log10(PIXEL_MAX / np.sqrt(mse))
```

```python
print(psnr(ref_image, img_tif))
print(psnr(ref_image, img_png))
print(psnr(ref_image, img_jpg10))

print(psnr(ref_image, np.random.randint(0, 255, ref_image.shape))) #compare any random image with a noise
```
